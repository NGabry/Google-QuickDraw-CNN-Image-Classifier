# Numpy bit map to folders jpg train / test sets 
Import numpy bitmaps of +10,000 28x28 images, and output them into split train and test folders seperated by image category


```python
import numpy as np
import matplotlib.pyplot as plt
import os 
from PIL import Image
%matplotlib inline
plt.style.use('dark_background')
```

```python
# Load a bit map, and do some reshape to make sure that it looks correct
airplane = np.load('drawings/numpy_bitmap_airplane.npy')
airplane = airplane.reshape(100,100,28,28)
airplane = airplane.transpose(0,2,1,3)
airplane = airplane.reshape(2800,2800)
plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(airplane[:280,:280],cmap='gray')
plt.show()
```

```python
categories =  ['airplane',
               'coffee_cup',
               'spoon', 
               'bicycle',
               'shoe']
```

```python
os.mkdir('data')
os.chdir('./data')
os.system('mkdir train test')
for i in categories:
    bitmap = np.load('../drawings/numpy_bitmap_'+i+'.npy')
    bitmap = bitmap.reshape(10000,28,28)
    os.mkdir('train/'+i)
    for j in range(8000):
        plt.imsave('train/'+i+'/'+str(j)+'.png',bitmap[j],cmap='gray')
    os.mkdir('test/'+i)
    for k in range(8001,10000,1):
        plt.imsave('test/'+i+'/'+str(k)+'.png',bitmap[j],cmap='gray')
```
